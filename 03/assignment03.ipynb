# <font color='orange'>Data preprocessing</font>

```python
#change column to category where it is appropriate
df = df.astype({"victory_status":"category", 
                "winner":"category", 
                "increment_code":"category", 
                "opening_eco":"category", 
                "opening_name":"category"})
```

```python
#since these colums have lost their float precision in the csv, they are quite useless to us
df.drop(columns=['created_at','last_move_at'],inplace=True)

#TODO decide if we should drop more columns
```

```python
#create columns with last move column, row and piece, to extract checkmate data more easily
def parse_last_move(moves):
    # parse the moves to extract the last move
    last_move = moves.split()[-1]
    col = row = piece = None
    
    #skips row if last move was not a checkmate
    if( last_move[-1] == '#' ):
        
        #if last move is a pawn promotion
        if ( '=' in last_move ):
            piece = last_move[-2]
            row = last_move[-5]
            col = last_move[-4]
        else :
            piece = last_move[0]
            if ( piece not in ['Q','R','N','B'] ):
                piece = 'P'
            row = last_move[-3]
            col = last_move[-2]
    
    return piece,row,col

# apply the function to the 'moves' column of the df
df['last_move_piece'],df['last_move_row'],df['last_move_col'] = zip(*df['moves'].map(parse_last_move))
```

```python
#extraction of taken chess pieces for each side 

def count_losses(board, piece, color):
        if piece == chess.QUEEN:
            piece_count = 1
        elif piece in (chess.ROOK, chess.BISHOP, chess.KNIGHT):
            piece_count = 2
        elif piece == chess.PAWN:
            piece_count = 8

        capture_count = piece_count - chess.popcount(board.pieces_mask(piece, color))

        return capture_count if capture_count >= 0 else 0

def extract_taken(moves):
    pgn = io.StringIO(moves)
    game = chess.pgn.read_game(pgn)
    board = game.end().board()
    white_captured = list()
    black_captured = list()
    
    return ( count_losses(board,chess.PAWN,chess.WHITE),
    count_losses(board,chess.KNIGHT,chess.WHITE),
    count_losses(board,chess.BISHOP,chess.WHITE),
    count_losses(board,chess.ROOK,chess.WHITE),
    count_losses(board,chess.QUEEN,chess.WHITE),
    count_losses(board,chess.PAWN,chess.BLACK),
    count_losses(board,chess.KNIGHT,chess.BLACK),
    count_losses(board,chess.BISHOP,chess.BLACK),
    count_losses(board,chess.ROOK,chess.BLACK),
    count_losses(board,chess.QUEEN,chess.BLACK) )
    

df['white_pawns_lost'],df['white_knights_lost'],df['white_bishops_lost'],df['white_rooks_lost'],df['white_queens_lost'],df['black_pawns_lost'],df['black_knights_lost'],df['black_bishops_lost'],df['black_rooks_lost'],df['black_queens_lost'] = zip(*df['moves'].map(extract_taken))
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,5))

# create a pivot table with the checkmate moves
pivot_table = df.pivot_table(values='last_move_piece', index='last_move_col', columns='last_move_row', aggfunc='count')

# create a heatmap using the pivot table
sns.heatmap(pivot_table, cmap='YlGnBu',ax=ax[0])

bars = sns.countplot(x=df['last_move_piece'],ax=ax[1],order = df['last_move_piece'].value_counts().index,palette='YlGnBu')
labels=["Queen","Rook","Bishop","Knight","Pawn"]
ax[1].set_xticklabels(labels)

ax[0].set_xlabel(None)
ax[0].set_ylabel(None)
ax[1].set_xlabel(None)

ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].spines['left'].set_visible(False)
ax[1].yaxis.set_visible(False)

ax[1].bar_label(ax[1].containers[0])


ax[0].set_title("Heatmap of Checkmate moves on the chess board")
ax[1].set_title("Number of checkmates based on the piece that delivered it")
plt.show()
```
